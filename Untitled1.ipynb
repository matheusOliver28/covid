```python
import matplotlib.pyplot as plt
import numpy as np

# Dados fornecidos para o ano de 2022
dados_regioes_2022 = {
    "Centro-Oeste": {"populacao": 15875907, "vacinados": 12259217},
    "Nordeste": {"populacao": 54644582, "vacinados": 42252908},
    "Norte": {"populacao": 16906962, "vacinados": 12140707},
    "Sudeste": {"populacao": 83800000, "vacinados": 70800022},
    "Sul": {"populacao": 29200000, "vacinados": 23932788}
}

# Listas para armazenar os valores de vacinados e não vacinados por região
regioes = list(dados_regioes_2022.keys())
vacinados = []
nao_vacinados = []

# Cálculo de vacinados e não vacinados por região
for regiao in regioes:
    vacinados_regiao = dados_regioes_2022[regiao]["vacinados"]
    populacao_regiao = dados_regioes_2022[regiao]["populacao"]
    nao_vacinados_regiao = populacao_regiao - vacinados_regiao
    vacinados.append(vacinados_regiao)
    nao_vacinados.append(nao_vacinados_regiao)

# Configuração do gráfico
x = np.arange(len(regioes))  # Posições das regiões no eixo x
largura = 0.35  # Largura das barras

# Criar gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))
barras_vacinados = ax.bar(x - largura/2, vacinados, largura, label='Vacinados', color='skyblue')
barras_nao_vacinados = ax.bar(x + largura/2, nao_vacinados, largura, label='Não Vacinados', color='salmon')

# Adicionar rótulos e título
ax.set_xlabel('Regiões')
ax.set_ylabel('População')
ax.set_title('Comparação entre Vacinados e Não Vacinados por Região em 2021')
ax.set_xticks(x)
ax.set_xticklabels(regioes)
ax.legend()

# Exibir os valores nas barras
for barra in barras_vacinados + barras_nao_vacinados:
    altura = barra.get_height()
    ax.annotate(f'{altura / 1_000_000:.1f}M',  # Mostrar em milhões para clareza
                xy=(barra.get_x() + barra.get_width() / 2, altura),
                xytext=(0, 3),  # Deslocamento acima da barra
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()  # Ajuste do layout
plt.show()
```

O código acima exibe os números de vacinados e não vacinados por região em 2021.

```python
import pandas as pd

# Dados dos não vacinados
dados_nao_vacinados = {
    'Ano': [2021, 2022, 2023],
    'Nao_Vacinados': [
        21168791 - 16478518,  # 2021
        21168791 - 18290289,  # 2022
        21168791 - 18575878   # 2023
    ]
}
df_nao_vacinados = pd.DataFrame(dados_nao_vacinados)

# Carregar os arquivos de cada ano e adicionar uma coluna de ano
arquivos = {
    2021: '/content/dosesMG2021.xlsx',
    2022: '/content/dosesMG2022.xlsx',
    2023: '/content/dosesMG2023.xlsx'
}

# Lista para armazenar os DataFrames de cada ano
df_vacinados_list = []

for ano, caminho in arquivos.items():
    # Carregar o arquivo do ano específico
    df_ano = pd.read_excel(caminho, sheet_name='Sheet1')

    # Adicionar coluna de ano e filtrar as colunas de interesse
    df_ano['Ano'] = ano
    df_ano = df_ano[['Ano', '1ª Dose', '2ª Dose']]

    # Adicionar o DataFrame à lista
    df_vacinados_list.append(df_ano)

# Combinar os DataFrames de cada ano em um único DataFrame
df_vacinados_total = pd.concat(df_vacinados_list, ignore_index=True)

# Mesclar com os dados de não vacinados com base no ano
df_combinado = pd.merge(df_nao_vacinados, df_vacinados_total, on='Ano', how='inner')

# Exibir o DataFrame combinado
print("DataFrame combinado com dados de não vacinados e vacinados por ano:")
print(df_combinado)
```

A taxa de óbito apresentada acima é um dado fornecido pelo Governo de Minas gerais, utilizado para calcular e comprovar a efetividade da vacina no estado comparando vacinados com a 1ª dose vs não vacinados.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fornecidos
dados_doses = {
    'Ano': [2021, 2022, 2023],
    'Populacao': [21168791, 21168791, 21168791],
    '1ª Dose': [16478318, 18290289, 18575878],
    '2ª Dose': [14818673, 16799391, 17098163]
}

# Taxas de óbito por 100.000 habitantes
taxa_obito_1_dose = 0.12
taxa_obito_2_dose = 0.06
taxa_obito_nao_vacinados = 0.71

# Criar DataFrame
df_doses = pd.DataFrame(dados_doses)

# Calcular número de não vacinados
df_doses['Não Vacinados'] = df_doses['Populacao'] - df_doses['1ª Dose']

# Calcular óbitos para cada grupo
df_doses['Óbitos 1ª Dose'] = (df_doses['1ª Dose'] * taxa_obito_1_dose) / 100_000
df_doses['Óbitos 2ª Dose'] = (df_doses['2ª Dose'] * taxa_obito_2_dose) / 100_000
df_doses['Óbitos Não Vacinados'] = (df_doses['Não Vacinados'] * taxa_obito_nao_vacinados) / 100_000

# Exibir o DataFrame com os cálculos de óbitos
print(df_doses[['Ano', 'Óbitos 1ª Dose', 'Óbitos 2ª Dose', 'Óbitos Não Vacinados']])

# Criar gráfico de comparação
anos = df_doses['Ano']
obitos_1_dose = df_doses['Óbitos 1ª Dose']
obitos_2_dose = df_doses['Óbitos 2ª Dose']
obitos_nao_vacinados = df_doses['Óbitos Não Vacinados']

# Configurar o gráfico
plt.figure(figsize=(10, 6))
bar_width = 0.2
x = range(len(anos))

# Barras para cada grupo de óbitos
plt.bar([i - bar_width for i in x], obitos_1_dose, width=bar_width, label='Óbitos 1ª Dose', color='orange')
plt.bar(x, obitos_2_dose, width=bar_width, label='Óbitos 2ª Dose', color='green')
plt.bar([i + bar_width for i in x], obitos_nao_vacinados, width=bar_width, label='Óbitos Não Vacinados', color='red')

# Adicionar rótulos e título
plt.xlabel('Ano')
plt.ylabel('Número de Óbitos (por 100 mil habitantes)')
plt.title('Comparação de Óbitos por Taxa de Vacinação em Minas Gerais (2021-2023)')
plt.xticks(x, anos)
plt.legend()

plt.tight_layout()
plt.show()
```

O gráfico traz a comparação do número de mortes de vacinados(1ªdose),(2ªdose) e não vacinados a cada 100 mil habitantes, comparação ao longo de 2021,2022,2023.(O cálculo do número do número de óbitos está sendo feio em cima do número total acumulado de vacinados da 1ªdose e da 2ª dose).


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados fornecidos
dados_doses = {
    'Ano': [2021, 2022, 2023],
    'Populacao': [21168791, 21168791, 21168791],
    '1ª Dose': [16478318, 18290289, 18575878],
    '2ª Dose': [14818673, 16799391, 17098163]
}

# Taxas de óbito por 100 mil habitantes
taxa_obito_1_dose = 0.12
taxa_obito_2_dose = 0.06
taxa_obito_nao_vacinados = 0.71

# Criar DataFrame
df_doses = pd.DataFrame(dados_doses)

# Calcular número de pessoas em cada categoria
df_doses['Apenas 1ª Dose'] = df_doses['1ª Dose'] - df_doses['2ª Dose']
df_doses['Ambas as Doses'] = df_doses['2ª Dose']
df_doses['Não Vacinados'] = df_doses['Populacao'] - df_doses['1ª Dose']

# Calcular número de óbitos para cada grupo
df_doses['Óbitos 1ª Dose'] = (df_doses['Apenas 1ª Dose'] * taxa_obito_1_dose) / 100_000
df_doses['Óbitos 2ª Dose'] = (df_doses['Ambas as Doses'] * taxa_obito_2_dose) / 100_000
df_doses['Óbitos Não Vacinados'] = (df_doses['Não Vacinados'] * taxa_obito_nao_vacinados) / 100_000

# Criar variáveis para a regressão linear (população em milhões)
df_doses['População 1ª Dose (milhões)'] = df_doses['Apenas 1ª Dose'] / 1_000_000
df_doses['População 2ª Dose (milhões)'] = df_doses['Ambas as Doses'] / 1_000_000
df_doses['População Não Vacinados (milhões)'] = df_doses['Não Vacinados'] / 1_000_000

# Realizar a regressão linear para cada grupo separadamente
# 1ª Dose
modelo_1_dose = LinearRegression().fit(df_doses[['População 1ª Dose (milhões)']], df_doses['Óbitos 1ª Dose'])
y_pred_1_dose = modelo_1_dose.predict(df_doses[['População 1ª Dose (milhões)']])

# 2ª Dose
modelo_2_dose = LinearRegression().fit(df_doses[['População 2ª Dose (milhões)']], df_doses['Óbitos 2ª Dose'])
y_pred_2_dose = modelo_2_dose.predict(df_doses[['População 2ª Dose (milhões)']])

# Não Vacinados
modelo_nao_vacinados = LinearRegression().fit(df_doses[['População Não Vacinados (milhões)']], df_doses['Óbitos Não Vacinados'])
y_pred_nao_vacinados = modelo_nao_vacinados.predict(df_doses[['População Não Vacinados (milhões)']])

# Plotar os resultados da regressão linear
plt.figure(figsize=(12, 6))

# Gráfico para 1ª Dose
plt.subplot(1, 3, 1)
plt.scatter(df_doses['População 1ª Dose (milhões)'], df_doses['Óbitos 1ª Dose'], color='orange', label='Óbitos 1ª Dose')
plt.plot(df_doses['População 1ª Dose (milhões)'], y_pred_1_dose, color='orange', linestyle='--')
plt.xlabel('População 1ª Dose (milhões)')
plt.ylabel('Óbitos Estimados(100 mil habitantes)')
plt.title('Regressão Linear - Óbitos 1ª Dose')

# Gráfico para 2ª Dose
plt.subplot(1, 3, 2)
plt.scatter(df_doses['População 2ª Dose (milhões)'], df_doses['Óbitos 2ª Dose'], color='blue', label='Óbitos 2ª Dose')
plt.plot(df_doses['População 2ª Dose (milhões)'], y_pred_2_dose, color='blue', linestyle='--')
plt.xlabel('População 2ª Dose (milhões)')
plt.title('Regressão Linear - Óbitos 2ª Dose')

# Gráfico para Não Vacinados
plt.subplot(1, 3, 3)
plt.scatter(df_doses['População Não Vacinados (milhões)'], df_doses['Óbitos Não Vacinados'], color='red', label='Óbitos Não Vacinados')
plt.plot(df_doses['População Não Vacinados (milhões)'], y_pred_nao_vacinados, color='red', linestyle='--')
plt.xlabel('População Não Vacinados (milhões)')
plt.title('Regressão Linear - Óbitos Não Vacinados')

# Ajustar layout e exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Dados fornecidos
dados_doses = {
    'Ano': [2021, 2022, 2023],
    'Populacao': [21168791, 21168791, 21168791],
    '1ª Dose': [16478318, 18290289, 18575878],
    '2ª Dose': [14818673, 16799391, 17098163]
}

# Taxas de óbito por 100 mil habitantes
taxa_obito_1_dose = 0.12
taxa_obito_2_dose = 0.06
taxa_obito_nao_vacinados = 0.71

# Criar DataFrame
df_doses = pd.DataFrame(dados_doses)

# Simulação: igualando o número de 1ª dose e ambas as doses aos não vacinados
df_doses['Não Vacinados'] = df_doses['Populacao'] - df_doses['1ª Dose']
df_doses['Apenas 1ª Dose Simulada'] = df_doses['Não Vacinados']  # Simulando que 1ª dose é igual aos não vacinados
df_doses['Ambas as Doses Simulada'] = df_doses['Não Vacinados']  # Simulando que ambas as doses é igual a não vacinados

# Calcular número de óbitos para cada grupo com as taxas fornecidas
df_doses['Óbitos 1ª Dose Simulada'] = (df_doses['Apenas 1ª Dose Simulada'] * taxa_obito_1_dose) / 100_000
df_doses['Óbitos 2ª Dose Simulada'] = (df_doses['Ambas as Doses Simulada'] * taxa_obito_2_dose) / 100_000
df_doses['Óbitos Não Vacinados Simulada'] = (df_doses['Não Vacinados'] * taxa_obito_nao_vacinados) / 100_000

# Criar variáveis para a regressão linear (população em milhões)
df_doses['População 1ª Dose Simulada (milhões)'] = df_doses['Apenas 1ª Dose Simulada'] / 1_000_000
df_doses['População 2ª Dose Simulada (milhões)'] = df_doses['Ambas as Doses Simulada'] / 1_000_000
df_doses['População Não Vacinados Simulada (milhões)'] = df_doses['Não Vacinados'] / 1_000_000

# Realizar a regressão linear para cada grupo separadamente
# 1ª Dose Simulada
modelo_1_dose = LinearRegression().fit(df_doses[['População 1ª Dose Simulada (milhões)']], df_doses['Óbitos 1ª Dose Simulada'])
y_pred_1_dose = modelo_1_dose.predict(df_doses[['População 1ª Dose Simulada (milhões)']])

# 2ª Dose Simulada
modelo_2_dose = LinearRegression().fit(df_doses[['População 2ª Dose Simulada (milhões)']], df_doses['Óbitos 2ª Dose Simulada'])
y_pred_2_dose = modelo_2_dose.predict(df_doses[['População 2ª Dose Simulada (milhões)']])

# Não Vacinados Simulados
modelo_nao_vacinados = LinearRegression().fit(df_doses[['População Não Vacinados Simulada (milhões)']], df_doses['Óbitos Não Vacinados Simulada'])
y_pred_nao_vacinados = modelo_nao_vacinados.predict(df_doses[['População Não Vacinados Simulada (milhões)']])

# Plotar os resultados da regressão linear no mesmo gráfico
plt.figure(figsize=(10, 6))

# Gráfico para 1ª Dose Simulada
plt.scatter(df_doses['População 1ª Dose Simulada (milhões)'], df_doses['Óbitos 1ª Dose Simulada'], color='orange', label='Óbitos 1ª Dose Simulada')
plt.plot(df_doses['População 1ª Dose Simulada (milhões)'], y_pred_1_dose, color='orange', linestyle='--', label='Regressão 1ª Dose Simulada')

# Gráfico para 2ª Dose Simulada
plt.scatter(df_doses['População 2ª Dose Simulada (milhões)'], df_doses['Óbitos 2ª Dose Simulada'], color='blue', label='Óbitos 2ª Dose Simulada')
plt.plot(df_doses['População 2ª Dose Simulada (milhões)'], y_pred_2_dose, color='blue', linestyle='--', label='Regressão 2ª Dose Simulada')

# Gráfico para Não Vacinados Simulados
plt.scatter(df_doses['População Não Vacinados Simulada (milhões)'], df_doses['Óbitos Não Vacinados Simulada'], color='red', label='Óbitos Não Vacinados Simulados')
plt.plot(df_doses['População Não Vacinados Simulada (milhões)'], y_pred_nao_vacinados, color='red', linestyle='--', label='Regressão Não Vacinados Simulados')

# Configurações do gráfico
plt.xlabel('População (milhões)')
plt.ylabel('Óbitos Estimados')
plt.title('Regressão Linear - Comparação de Óbitos por Grupo (1ª Dose, 2ª Dose, Não Vacinados)')
plt.legend()

# Ajustar layout e exibir o gráfico
plt.tight_layout()
plt.show()
```

Para efeito comparativo simulamos a comparação do número de mortos(por 100 mil habitantes) entre não vacinados, vacinados(1ª dose) e (2ª dose) simulando ambos começando apartir de um mesmo período e terminando em um mesmo período, simulando uma população do mesmo tamanho para ambos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do arquivo Excel
df = pd.read_excel('/content/faixa_etaria.xlsx')

# Visualizar todos os dados do DataFrame
print("Dados carregados:")
print(df)

# Agrupar os dados por Faixa Etária e tipo de Dose
df_grouped = df.groupby(['Faixa Etária', 'Doses'])['Total de Doses Aplicadas'].sum().unstack().fillna(0)

# Configurar a largura das barras
largura = 0.8

# Criar o gráfico
fig, ax = plt.subplots(figsize=(10, 6))

# Criar gráfico empilhado
df_grouped.plot(kind='bar', stacked=True, ax=ax, color=plt.cm.tab10.colors)

# Adicionar rótulos e título
ax.set_xlabel('Faixa Etária', fontsize=12)
ax.set_ylabel('Total de Doses Aplicadas', fontsize=12)
ax.set_title('Total de Doses Aplicadas por Faixa Etária (tipos de doses)', fontsize=14)

# Adicionar legendas e ajustar sua posição
ax.legend(title='Tipo de Dose', bbox_to_anchor=(1.05, 1), loc='upper left')

# Ajustar layout para dar espaço à legenda
plt.tight_layout(pad=2.0)
```

O código acima exibe os dados sobre vacinas, organiza esses dados por faixa etária e tipo de dose, e apresenta essa informação de forma visual através de um gráfico.
