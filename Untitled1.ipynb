```python

import pandas as pd
import matplotlib.pyplot as plt

# Caminhos dos arquivos
caminho_obitos = '/content/obitos_covid.xlsx'
caminho_vacinacao = '/content/vacinacao_covid.xlsx'

def calcular_correlacao():
    try:
        # Carregar os dados de óbitos e vacinação
        df_obitos = pd.read_excel(caminho_obitos)
        df_vacinacao = pd.read_excel(caminho_vacinacao)

        # Listar as colunas para verificar os nomes
        print("Colunas em obitosGeral.xlsx:")
        print(df_obitos.columns)
        print("\nColunas em vacinacaoGeral.xlsx:")
        print(df_vacinacao.columns)

        # Renomear as colunas para facilitar o acesso
        df_obitos.rename(columns={
            'Taxa mortalidade (100 mil hab)': 'taxa_mortalidade',
            'Óbitos Acumulados': 'total_obitos'
        }, inplace=True)

        df_vacinacao.rename(columns={
            'Total de Doses Aplicadas': 'total_doses'
        }, inplace=True)

        # Normalizar os nomes dos estados/UF
        df_obitos['UF'] = df_obitos['UF'].str.strip().str.lower()
        df_vacinacao['UF'] = df_vacinacao['UF'].str.strip().str.lower()

        # Unir os DataFrames com base na UF
        df = pd.merge(df_obitos, df_vacinacao, on='UF', how='inner')

        # Verificando valores ausentes
        print(f"\nDados após merge:")
        print(df.head())
        print("\nVerificando valores ausentes:")
        print(df.isnull().sum())

        # Remover linhas com valores ausentes
        df_cleaned = df.dropna(subset=['taxa_mortalidade', 'total_doses'])

        # Calcular a correlação entre taxa de mortalidade e total de doses
        if len(df_cleaned) > 1:  # Necessário pelo menos 2 pontos para calcular correlação
            correlacao = df_cleaned['taxa_mortalidade'].corr(df_cleaned['total_doses'])
            print(f'A correlação entre taxa de mortalidade e total de doses aplicadas é: {round(correlacao, 2):.2f}')
        else:
            print("Não há dados suficientes para calcular a correlação.")

        # Plotar o gráfico de dispersão
        plt.figure(figsize=(10, 6))
        plt.scatter(df_cleaned['total_doses'], df_cleaned['taxa_mortalidade'], color='blue', alpha=0.5)
        plt.title('Gráfico de Dispersão: Taxa de Mortalidade vs Total de Doses Aplicadas')
        plt.xlabel('Total de Doses Aplicadas')
        plt.ylabel('Taxa de Mortalidade (100 mil hab)')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.tight_layout()  # Ajusta o layout para evitar sobreposição
        plt.show()

    except FileNotFoundError as e:
        print(f"Erro: {e}")

# Calcular a correlação
calcular_correlacao()





```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Caminhos dos arquivos
caminho_obitos = '/content/obitos_covid.xlsx'
caminho_vacinacao = '/content/vacinacao_covid.xlsx'

def calcular_correlacao():
    try:
        # Carregar os dados de óbitos e vacinação
        df_obitos = pd.read_excel(caminho_obitos)
        df_vacinacao = pd.read_excel(caminho_vacinacao)

        # Listar as colunas para verificar os nomes
        print("Colunas em obitos_covid.xlsx:")
        print(df_obitos.columns)
        print("\nColunas em vacinacao_covid.xlsx:")
        print(df_vacinacao.columns)

        # Renomear as colunas para facilitar o acesso
        df_obitos.rename(columns={
            'Casos Acumulados': 'total_casos',
            'Óbitos Acumulados': 'total_obitos'
        }, inplace=True)

        df_vacinacao.rename(columns={
            'Total de Doses Aplicadas': 'total_doses'
        }, inplace=True)

        # Normalizar os nomes dos estados/UF
        df_obitos['UF'] = df_obitos['UF'].str.strip().str.lower()
        df_vacinacao['UF'] = df_vacinacao['UF'].str.strip().str.lower()

        # Unir os DataFrames com base na UF
        df = pd.merge(df_obitos, df_vacinacao, on='UF', how='inner')

        # Verificando valores ausentes
        print(f"\nDados após merge:")
        print(df.head())
        print("\nVerificando valores ausentes:")
        print(df.isnull().sum())

        # Remover linhas com valores ausentes
        df_cleaned = df.dropna(subset=['total_casos', 'total_obitos'])

        # Calcular a correlação entre total de casos e total de óbitos
        if len(df_cleaned) > 1:  # Necessário pelo menos 2 pontos para calcular correlação
            correlacao = df_cleaned['total_casos'].corr(df_cleaned['total_obitos'])
            print(f'A correlação entre o total de casos e o total de óbitos é: {round(correlacao, 2):.2f}')
        else:
            print("Não há dados suficientes para calcular a correlação.")

        # Plotar o gráfico de dispersão
        plt.figure(figsize=(10, 6))
        plt.scatter(df_cleaned['total_obitos'], df_cleaned['total_casos'], color='blue', alpha=0.5)

        # Calcular a linha de regressão
        slope, intercept, r_value, p_value, std_err = stats.linregress(df_cleaned['total_obitos'], df_cleaned['total_casos'])

        # Gerar dados para a linha de regressão
        regression_line = slope * df_cleaned['total_obitos'] + intercept

        # Plotar a linha de regressão
        plt.plot(df_cleaned['total_obitos'], regression_line, color='red', linewidth=2, label='Linha de Regressão')

        plt.title('Gráfico de Dispersão: Total de casos vs Total de Óbitos')
        plt.xlabel('Total de Óbitos')
        plt.ylabel('Total de Casos')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.legend()  # Adicionar legenda
        plt.tight_layout()  # Ajusta o layout para evitar sobreposição
        plt.show()

    except FileNotFoundError as e:
        print(f"Erro: {e}")

# Calcular a correlação
calcular_correlacao()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# Caminhos dos arquivos
caminho_obitos = '/content/obitos_covid.xlsx'
caminho_vacinacao = '/content/vacinacao_covid.xlsx'

def calcular_correlacao():
    try:
        # Carregar os dados de óbitos e vacinação
        df_obitos = pd.read_excel(caminho_obitos)
        df_vacinacao = pd.read_excel(caminho_vacinacao)

        # Listar as colunas
        print("Colunas em obitos_covid.xlsx:")
        print(df_obitos.columns.tolist())
        print("\nColunas em vacinacao_covid.xlsx:")
        print(df_vacinacao.columns.tolist())

        # Renomear colunas
        df_obitos.rename(columns={
            'Casos novos notificados na semana epidemiológica': 'total_casos_novos_na_semana',
            'Óbitos Acumulados': 'total_obitos'
        }, inplace=True)

        df_vacinacao.rename(columns={
            'Total de Doses Aplicadas': 'total_doses'
        }, inplace=True)

        # Normalizar nomes das UF
        df_obitos['UF'] = df_obitos['UF'].str.strip().str.lower()
        df_vacinacao['UF'] = df_vacinacao['UF'].str.strip().str.lower()

        # Merge
        df = pd.merge(df_obitos, df_vacinacao, on='UF', how='inner')

        # Verificação
        print(f"\nDados após merge:")
        print(df.head())
        print("\nVerificando valores ausentes:")
        print(df.isnull().sum())

        # Limpeza de dados
        print(f"Linhas antes da limpeza: {len(df)}")
        df_cleaned = df.dropna(subset=['total_casos_novos_na_semana', 'total_doses'])
        print(f"Linhas após a limpeza: {len(df_cleaned)}")

        # Correlação
        if len(df_cleaned) > 1:
            correlacao = df_cleaned['total_casos_novos_na_semana'].corr(df_cleaned['total_doses'])
            print(f'A correlação entre o total de casos na semana e o total de doses aplicadas é: {round(correlacao, 2)}')
        else:
            print("Não há dados suficientes para calcular a correlação.")

        # Gráfico de dispersão
        plt.figure(figsize=(10, 6))
        plt.scatter(df_cleaned['total_doses'], df_cleaned['total_casos_novos_na_semana'], color='blue', alpha=0.5)

        # Regressão
        slope, intercept, r_value, p_value, std_err = stats.linregress(df_cleaned['total_doses'], df_cleaned['total_casos_novos_na_semana'])
        regression_line = slope * df_cleaned['total_doses'] + intercept

        plt.plot(df_cleaned['total_doses'], regression_line, color='red', linewidth=2, label='Linha de Regressão')
        plt.title('Gráfico de Dispersão: Total de casos novos na semana epidemiológica vs Total de Doses')
        plt.xlabel('Total de Doses')
        plt.ylabel('Total de Casos na semana epidemiológica')
        plt.grid(True)
        plt.xticks(rotation=45)
        plt.legend()
        plt.tight_layout()
        plt.show()

    except FileNotFoundError as e:
        print(f"Erro: {e}")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

# Calcular a correlação
calcular_correlacao()
```

O aumento do número de casos está fortemente relacionado ao aumento no número de doses aplicadas.
A vacinação está acompanhado o aumento da transmissão do vírus.

Interpretações possíveis:
1-A vacinação está sendo direcionada para áreas com alto número de casos.
2-A populção está procurando vacinação em resposta ao aumento de casos.
3-A vacinação está ajudando a controlar a transmissão do vírus.
